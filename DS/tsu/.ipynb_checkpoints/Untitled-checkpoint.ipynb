```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
plt.plot(x, y)

plt.title("sin(x)")
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```
