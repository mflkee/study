### Задание 1

Выполнить импорт библиотеки *NumPy*, модуля *Pyplot* библиотеки *Matplotlib* и модуля *stats* библиотеки *SciPy*.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

Случайная величина $X$ задана функцией плотности распределения вероятностей
$$
f(x)=
\begin{cases}
0 &  x<0, \\
1{,}5 \, e^{-1{,}5 x} & x \ge 0.
\end{cases}
$$


Определить  класс случайной величины $X$ (дискретная или непрерывная), ее закон распределения и значение параметров закона распределения.   
**Указание**: описание основных законов распределения дано в презентации *Основные законы распределения*.



*Ваш ответ:*

Случайная величина **непрерывная**. По виду плотности
$f(x)=\lambda e^{-\lambda x},\ x\ge 0$ это **экспоненциальное распределение**.
Параметр (интенсивность) $\lambda=1{,}5$ (масштаб $\theta=1/\lambda=2/3$).


Используя соответствующий класс модуля *stats*, создать случайную величину с данным распределением.

```python
# Экспоненциальное распределение с параметром λ=1.5 (scale = 1/λ)
rv_221 = stats.expon(loc=0, scale=1/1.5)
```

Для случайной величины *rv_221* построить график функции
плотности распределения вероятностей. Выставить метки по осям: для оси абсцисс – значения, начиная с 0, c шагом, равным 0,5, для оси ординат – значения, начиная с 0, c шагом, равным 0,1. Добавить сетку командой `plt.grid()`.

**Указание**: подберите левую и правую границы массива абсцисс точек так, чтобы на графике были построены обе ветви функции плотности распределения и чтобы было видно стремление функции к 0 при $x\rightarrow +\infty$. Количество точек в массиве должно быть достаточно большим: линии разрыва не должны быть наклонными.    



```python
# График плотности распределения
x = np.linspace(-2, 6, 2001)  # достаточно много точек, чтобы разрыв в 0 был вертикальным
y = rv_221.pdf(x)

plt.figure()
plt.plot(x, y)
plt.xticks(np.arange(0, 6.5, 0.5))
plt.yticks(np.arange(0, 1.6, 0.1))
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Плотность экспоненциального распределения (λ=1.5)')
plt.show()
```

По графику плотности распределения определить максимальное значение функции плотности распределения. Как оно связано со значением параметра распределения?

*Ваш ответ:*

Максимум плотности достигается при $x=0$ и равен $f(0)=\lambda=1{,}5$.
То есть максимальное значение плотности **совпадает со значением параметра интенсивности** $\lambda$.

Для случайной величины *rv_221* построить график функции распределения. Добавить сетку командой `plt.grid()`.

**Указание**: используйте тот же массив абсцисс точек, что и для функции плотности распределения.

```python
# График функции распределения (CDF)
plt.figure()
plt.plot(x, rv_221.cdf(x))
plt.xticks(np.arange(0, 6.5, 0.5))
plt.yticks(np.arange(0, 1.05, 0.1))
plt.grid()
plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('Функция распределения экспоненциального распределения (λ=1.5)')
plt.show()
```

Согласно графику функции распределения, к какому значению стремится функция распределения при $x\rightarrow +\infty$?

*Ваш ответ:*

$F(x)\to 1$ при $x\to +\infty$ (как и у любой функции распределения непрерывной СВ с полной вероятностью 1).

### Задание 2

При работе некоторого устройства время от времени возникают сбои. Случайная величина $X$ – количество сбоев за сутки, – имеет распределение Пуассона. Среднее количество сбоев за сутки равно 3.

Определить  класс случайной величины $X$, ее закон распределения и значение параметров закона распределения.   
**Указание**: описание основных законов распределения дано в презентации *Основные законы распределения*.



*Ваш ответ:*

Случайная величина **дискретная**. Закон распределения — **распределение Пуассона**.
Параметр $\lambda$ равен математическому ожиданию: $\lambda=3$.


Используя соответствующий класс модуля *stats*, создать случайную величину с данным распределением.

```python
# Распределение Пуассона с параметром λ = 3
failures = stats.poisson(mu=3)
```

Для случайной величины *failures* построить многоугольник распределения. Выставить метки по осям: для оси абсцисс использовать созданный массив абсцисс точек, для оси ординат – значения, начиная с 0, c шагом, равным 0,01. Добавить сетку командой `plt.grid()`.

**Указание**: массив абcцисс точек должен содержать только возможные значения случайной величины. Опытным путем подберите правую границу массива так, чтобы было видно стремление вероятности к 0.

```python
# Многоугольник распределения (PMF)
k = np.arange(0, 16)  # возможные значения 0..15, далее вероятность уже очень мала
pmf = failures.pmf(k)

plt.figure()
plt.plot(k, pmf, marker='o')
plt.xticks(k)
plt.yticks(np.arange(0, max(pmf) + 0.01, 0.01))
plt.grid()
plt.xlabel('k')
plt.ylabel('P(X=k)')
plt.title('Многоугольник распределения Пуассона (λ=3)')
plt.show()
```

По многоугольнику распределения определить примерное значение вероятности одного cбоя за сутки.

*Ваш ответ:*

$P(X=1)=\dfrac{3^1 e^{-3}}{1!}=3e^{-3}\approx 0{,}149$ (около 0.15).

Для случайной величины *failures* построить график функции распределения. Выставить метки по осям: для оси абсцисс использовать созданный массив абсцисс точек, для оси ординат – значения на отрезке [0, 1] c шагом, равным 0,05.  Добавить сетку командой `plt.grid()`.

**Указание**: подберите левую и правую границы массива абсцисс точек так, чтобы на графике было представлено минимальное из возможных значений функции распределения и чтобы было видно стремление функции к 1 при $x\rightarrow +\infty$.

```python
# Функция распределения (CDF) для дискретной СВ удобно рисовать ступенчатой
k_cdf = np.arange(0, 16)
cdf_vals = failures.cdf(k_cdf)

plt.figure()
plt.step(k_cdf, cdf_vals, where='post')
plt.xticks(k_cdf)
plt.yticks(np.arange(0, 1.05, 0.05))
plt.grid()
plt.xlabel('k')
plt.ylabel('F(k)')
plt.title('Функция распределения Пуассона (λ=3)')
plt.show()
```

По графику функции распределения определить примерное значение вероятности того, что количество сбоев за сутки будет меньше 4, обосновав свой ответ.

*Ваш ответ c пояснениями:*

Нужно $P(X<4)=P(X\le 3)$. По графику функции распределения это значение $F(3)$.
Для $\lambda=3$ получаем $F(3)\approx 0{,}647$ (то есть примерно 0.65).

Для случайной величины *failures* вычислить математическое ожидание, дисперсию и среднее квадратическое отклонение. Если результат является бесконечной десятичной дробью, округлить его до 3 знаков после запятой.


```python
# Математическое ожидание, дисперсия и СКО
m = failures.mean()
d = failures.var()
s = failures.std()

print("E[X] =", round(m, 3))
print("D[X] =", round(d, 3))
print("σ =", round(s, 3))
```

Определить значение, которое количество сбоев за сутки не будет превосходить с вероятностью 0,5.

```python
# Медиана/квантиль уровня 0.5: минимальное k, для которого F(k) ≥ 0.5
q_05 = int(failures.ppf(0.5))
print("Значение, которое X не превосходит с вероятностью 0.5:", q_05)
```

### Задание 3

Случайная величина $X$ задана функцией распределения

$$
F(x) =
\begin{cases}
0, & x \le 8\\
\dfrac{x-8}{8}, & 8 < x \le 16\\
1, & x > 16
\end{cases}
$$

Определить  класс случайной величины $X$, ее закон распределения и значение параметров закона распределения.   
**Указание**: описание основных законов распределения дано в презентации *Основные законы распределения*.



*Ваш ответ:*

Случайная величина **непрерывная**. По виду функции распределения это **равномерное распределение** на отрезке $[8,16]$:
$X\sim U(8,16)$ (параметры: $a=8$, $b=16$).


Используя соответствующий класс модуля *stats*, создать случайную величину с данным распределением.

```python
# Равномерное распределение на [8, 16]
rv_223 = stats.uniform(loc=8, scale=8)
```

Для случайной величины *rv_223* построить график функции распределения. Выставить метки по осям: для оси абсцисс – целые значения c шагом, равным 1, для оси ординат – значения на отрезке [0, 1] c шагом, равным 0,05. Добавить сетку командой `plt.grid()`.

**Указание**: подберите левую и правую границы массива абсцисс точек так, чтобы на графике были построены все три ветви функции распределения. Количество точек в массиве должно быть достаточно большим.

```python
# График функции распределения (CDF)
x3 = np.linspace(6, 18, 2001)
plt.figure()
plt.plot(x3, rv_223.cdf(x3))
plt.xticks(np.arange(6, 19, 1))
plt.yticks(np.arange(0, 1.05, 0.05))
plt.grid()
plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('Функция распределения U(8,16)')
plt.show()
```

По графику функции распределения определить приблизительное значение вероятности того, что случайная величина $X$ примет значение из интервала (10,12), обосновав свой ответ.


*Ваш ответ с пояснениями:*

Для равномерного распределения на $[8,16]$:
\[
P(10<X<12)=\frac{12-10}{16-8}=\frac{2}{8}=0{,}25.
\]
На графике это разность значений $F(12)-F(10)$.

Для случайной величины  *rv_223* построить график функции плотности распределения вероятностей. Выставить метки по оси абсцисс – значения c шагом, равным 1. Добавить сетку командой `plt.grid()`.

**Указание**: используйте тот же массив абсцисс точек, что и для функции распределения. Количество точек в массиве должно быть достаточно большим: линии разрыва не должны быть наклонными.

```python
# График плотности (PDF)
plt.figure()
plt.plot(x3, rv_223.pdf(x3))
plt.xticks(np.arange(6, 19, 1))
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Плотность U(8,16)')
plt.show()
```

По графику плотности распределения определить, чему равен третий квартиль, обосновав свой ответ.

*Ваш ответ с пояснениями:*

Третий квартиль — это квантиль уровня 0.75.
Для $U(8,16)$: $Q_{0.75}=8+0.75\cdot(16-8)=8+6=14$.
На графике это точка, где $F(x)=0.75$.

Для случайной величины *rv_223* вычислить математическое ожидание, дисперсию и среднее квадратическое отклонение. Если результат является бесконечной десятичной дробью, округлить его до 2 знаков после запятой.

```python
# Математическое ожидание, дисперсия и СКО
m3 = rv_223.mean()
d3 = rv_223.var()
s3 = rv_223.std()

print("E[X] =", round(m3, 2))
print("D[X] =", round(d3, 2))
print("σ =", round(s3, 2))
```

### Задание 4

Текущая цена акции представляет собой нормально распределенную случайную величину $X$ со средней ценой 100 у.е. и дисперсией 225 у.е.²



Определить  класс случайной величины $X$, ее закон распределения и значение параметров закона распределения.   
**Указание**: описание основных законов распределения дано в презентации *Основные законы распределения*.



*Ваш ответ:*

Случайная величина **непрерывная**. Закон распределения — **нормальное распределение**.
$\mu=100$, $\sigma^2=225$, значит $\sigma=15$.
Итого: $X\sim \mathcal{N}(100,15^2)$.


Используя соответствующий класс модуля *stats*, создать случайную величину с данным распределением.

```python
# Нормальное распределение: μ=100, σ^2=225 => σ=15
price = stats.norm(loc=100, scale=15)
```

Для случайной величины *price* построить график функции распределения. Выставить метки по осям: для оси абсцисс – целые значения, кратные 10, для оси ординат – значения на отрезке [0, 1] c шагом, равным 0,25. Добавить сетку командой `plt.grid()`.

**Указание**: подберите левую и правую границы массива абсцисс точек так, чтобы на графике было видно стремление функции распределения к 0 при $x\rightarrow -\infty$ и к 1 при $x\rightarrow +\infty$. Количество точек в массиве должно быть достаточно большим.

```python
# График функции распределения (CDF)
x4 = np.linspace(40, 160, 2001)
plt.figure()
plt.plot(x4, price.cdf(x4))
plt.xticks(np.arange(40, 161, 10))
plt.yticks(np.arange(0, 1.01, 0.25))
plt.grid()
plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('Функция распределения N(100,15²)')
plt.show()
```

По графику функции распределения определить значение, которого текущая цена акции в 25% случаев не превышает, а в оставшихся 75% случаев цена будет больше этого числа, обосновав свой ответ.

*Ваш ответ c пояснениями:*

Искомое число — это **25-й перцентиль** (квантиль уровня 0.25): $x_{0.25}$, где $F(x_{0.25})=0.25$.
Для $\mathcal{N}(100,15^2)$:
\[
x_{0.25}=100+15\cdot z_{0.25},\quad z_{0.25}\approx -0.6745.
\]
Отсюда $x_{0.25}\approx 89.88$ (≈ 89.9).

Для случайной величины *price* построить график функции
плотности распределения вероятностей. Добавить сетку командой `plt.grid()`.

**Указание**: используйте тот же массив абсцисс точек, что и для функции распределения.

```python
# График плотности (PDF)
plt.figure()
plt.plot(x4, price.pdf(x4))
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Плотность N(100,15²)')
plt.show()
```

По графикам определить примерные значения левой и правой границ промежутка, в который случайная величина $X$ попадает с вероятностью, близкой к 1.

*Ваш ответ:*

По графикам (и по правилу «трёх сигма») практически вся масса распределения лежит примерно в интервале
$[m-3\sigma,\ m+3\sigma]=[100-45,\ 100+45]=[55,\ 145]$.



Согласно правилу «трех сигма» вероятность того, что случайная величина, распределенная нормально с параметрами $m$ и $\sigma$, отклонится от своего математического ожидания не более, чем на $3\sigma$, равна 0,9973. Проверьте это на практике.    

Найти границы $m-3\sigma$ и $m+3\sigma$ для данной случайной величины и вычислить вероятность $P(m-3\sigma < X < m+3\sigma)$, округлив полученное значение до 4 знаков после запятой.

```python
m4 = price.mean()
sigma4 = price.std()
left = m4 - 3*sigma4
right = m4 + 3*sigma4

prob_3sigma = price.cdf(right) - price.cdf(left)

print("m-3σ =", left)
print("m+3σ =", right)
print("P(m-3σ < X < m+3σ) =", round(prob_3sigma, 4))
```

### Задание 5

Вероятность случайной величине $X$ принять значение $𝑘 = 0, 1, 2, \dots, 16$
определяется по формуле $$P(X = k) = C^{k}_{16} \cdot 0{,}35^k \cdot 0{,}65^{16-k}$$


Определить  класс случайной величины $X$, ее закон распределения и значение параметров закона распределения.   
**Указание**: описание основных законов распределения дано в презентации *Основные законы распределения*.



*Ваш ответ:*

Случайная величина **дискретная**. По формуле это **биномиальное распределение**:
$X\sim Bin(n=16,\ p=0{,}35)$.


Используя соответствующий класс модуля *stats*, создать случайную величину с данным распределением.

```python
# Биномиальное распределение: n=16, p=0.35
rv_225 = stats.binom(n=16, p=0.35)
```

Для случайной величины *rv_225* построить многоугольник распределения. Выставить метки по осям: для оси абсцисс использовать созданный массив абсцисс точек, для оси ординат – значения, начиная с 0, c шагом, равным 0,05. Добавить сетку командой `plt.grid()`.

**Указание**: массив абсцисс точек должен содержать все возможные значения случайной величины и только их.

```python
# Многоугольник распределения (PMF)
k5 = np.arange(0, 17)
pmf5 = rv_225.pmf(k5)

plt.figure()
plt.plot(k5, pmf5, marker='o')
plt.xticks(k5)
plt.yticks(np.arange(0, max(pmf5) + 0.05, 0.05))
plt.grid()
plt.xlabel('k')
plt.ylabel('P(X=k)')
plt.title('Многоугольник распределения Bin(16, 0.35)')
plt.show()

# Вероятность в наиболее вероятной точке (k=5)
mode5 = int(np.floor((16 + 1) * 0.35))
print("Мода:", mode5)
print("P(X=mode) =", round(rv_225.pmf(mode5), 5))
```

### Задание 6

Пусть независимые случайные величины $X_1, X_2, \dots , X_{25}$ имеют одно и то же распределение с математическим ожиданием, равным 40, и дисперсией, равной 144.

Согласно центральной предельной теореме, случайная величина $$\overline{X} = \dfrac {1}{25} \sum_{k=1}^{25} X_k$$
будет иметь асимптотически нормальное распределение с математическим ожиданием $m$ и средним квадратическим отклонением $\sigma$.

Найти значения $m$ и $\sigma$, обосновав свой ответ.

*Ваш ответ с пояснениями:*

Для среднего $\overline{X}$:
- математическое ожидание $m=E(\overline{X})=E(X_k)=40$;
- дисперсия $D(\overline{X})=\dfrac{D(X_k)}{n}=\dfrac{144}{25}=5.76$;
- значит $\sigma=\sqrt{5.76}=2.4$.

Итого: $\overline{X}$ асимптотически $\mathcal{N}(40,\ 2.4^2)$.
