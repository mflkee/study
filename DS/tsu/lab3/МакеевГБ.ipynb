Используя соответствующий класс модуля *stats*, создать случайную величину с данным распределением.

По многоугольнику распределения определить примерное значение вероятности одного cбоя за сутки.

*Ваш ответ:*


```python

k_cdf = np.arange(0, 16)
cdf_vals = failures.cdf(k_cdf)

plt.figure()
plt.step(k_cdf, cdf_vals, where='post')
plt.xticks(k_cdf)
plt.yticks(np.arange(0, 1.05, 0.05))
plt.grid()
plt.xlabel('k')
plt.ylabel('F(k)')
plt.title('Функция распределения Пуассона (λ=3)')
plt.show()
```

По графику функции распределения определить примерное значение вероятности того, что количество сбоев за сутки будет меньше 4, обосновав свой ответ.

*Ваш ответ c пояснениями:*

Нам нужно значение меньше 4. По графику функции распределения это значение F(3) = 0.65.

Для случайной величины *failures* вычислить математическое ожидание, дисперсию и среднее квадратическое отклонение. Если результат является бесконечной десятичной дробью, округлить его до 3 знаков после запятой.

*Ваш ответ:*

Найти границы $m-3\sigma$ и $m+3\sigma$ для данной случайной величины и вычислить вероятность $P(m-3\sigma < X < m+3\sigma)$, округлив полученное значение до 4 знаков после запятой.

### Задание 6

Пусть независимые случайные величины $X_1, X_2, \dots , X_{25}$ имеют одно и то же распределение с математическим ожиданием, равным 40, и дисперсией, равной 144.

Согласно центральной предельной теореме, случайная величина $$\overline{X} = \dfrac {1}{25} \sum_{k=1}^{25} X_k$$
будет иметь асимптотически нормальное распределение с математическим ожиданием $m$ и средним квадратическим отклонением $\sigma$.

Найти значения $m$ и $\sigma$, обосновав свой ответ.

*Ваш ответ с пояснениями:*

Для среднего $\overline{X}$:
- математическое ожидание $m=E(\overline{X})=E(X_k)=40$;
- дисперсия $D(\overline{X})=\dfrac{D(X_k)}{n}=\dfrac{144}{25}=5.76$;
- значит $\sigma=\sqrt{5.76}=2.4$.

Итого: $\overline{X}$ асимптотически $\mathcal{N}(40,\ 2.4^2)$.
