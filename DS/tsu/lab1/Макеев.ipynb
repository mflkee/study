### Задание 8

Построить график функции  $$f(x) = \frac{2x-1}{(x-1)^2}$$ на отрезке [-1, 0.4] c шагом, равным 0.05.

Выставить метки по осям: по оси абсцисс с шагом 0.2; по оси ординат - с шагом 0.1.

```python
# Создаем массив x от -1 до 0.4 с шагом 0.05
x = np.arange(-1, 0.4 + 0.05, 0.05)

# Вычисляем значения функции f(x) = (2x-1)/((x-1)^2)
# Используем .T для корректных вычислений
f = (2*x - 1) / ((x - 1)**2)

# Построение графика
plt.figure(figsize=(12, 8))
plt.plot(x, f, 'b-', linewidth=2, marker='o', markersize=4, label='f(x) = (2x-1)/((x-1)^2)')

# Настройка меток по осям
plt.xticks(np.arange(-1, 0.5, 0.2))
plt.yticks(np.arange(-2, 0.5, 0.1))

# Добавление сетки, легенды и подписей
plt.grid(True, alpha=0.3, linestyle='--')
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.title('График функции f(x) = (2x-1)/((x-1)^2)', fontsize=14)
plt.legend(fontsize=11)

# Добавляем горизонтальную и вертикальную линии для наглядности
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)

# Отображение графика
plt.tight_layout()
plt.show()

# Определение ординаты точки локального минимума
min_index = np.argmin(f)
print(f"Локальный минимум функции:")
print(f"x = {x[min_index]:.2f}")
print(f"f(x) = {f[min_index]:.3f}")
print(f"Ордината точки локального минимума: {f[min_index]:.3f}")
```
