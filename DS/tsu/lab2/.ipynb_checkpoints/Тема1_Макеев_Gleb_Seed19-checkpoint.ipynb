После этого **все расчёты будут воспроизводимы**.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

# =========================
# 1) Заполните свои данные:
# =========================
SURNAME = "Макеев"   # <-- замените на вашу фамилию латиницей или кириллицей
SEED = 19             # <-- замените на ваш номер по списку в Modeus

# ==========================================================
# 2) ВАЖНО: замените на свою функцию из задания 8 (ваш вариант)
#    Пример ниже — ЗАГЛУШКА, его нельзя сдавать как решение!
# ==========================================================
def f(x: float) -> float:
    # Пример-заглушка с разрывом в x=1 (вертикальная асимптота)
    return (x**2 - 4*x + 5) / (x - 1)

# Производная:
# - Если ваша функция сложная, можно:
#   (a) выписать производную вручную и реализовать здесь,
#   (b) или использовать численную производную (ниже есть вариант).
def df(x: float) -> float:
    # Производная для заглушки:
    return ((2*x - 4)*(x - 1) - (x**2 - 4*x + 5)) / (x - 1)**2

print("Параметры заданы:", SURNAME, SEED)
```

## Задание 1. Исследование матричного уравнения  \(A \cdot X = B\)

**Что делаем:**
1. Генерируем матрицу `A` (3×3) и вектор `B` (3×1) со случайными элементами при одном и том же `seed`.
2. Находим определитель `det(A)` и делаем вывод о разрешимости.
3. Если `A` невырождена (`det(A) ≠ 0`), решаем уравнение и проверяем.

### Напишите своими словами (обязательно)
- Почему по определителю можно судить о наличии/единственности решения?
- Что означает «вырожденная матрица» в контексте уравнения \(A X = B\)?


```python
rng = np.random.default_rng(SEED)

A = rng.random((3, 3))
B = rng.random((3, 1))

A, B
```

```python
detA = np.linalg.det(A)
detA
```

### Вывод по определителю (как студент 2 курса)

- Мы вычислили `det(A)` — определитель квадратной матрицы.  
- Если `det(A) ≠ 0`, то матрица **невырождена** и имеет обратную `A^{-1}`. Тогда уравнение \(A X = B\) имеет **единственное** решение: \(X = A^{-1}B\).  
- Если `det(A) = 0` (или численно очень близок к 0), то матрица **вырождена**, обратной не существует, и тогда:
  - либо решений **нет** (если вектор `B` не лежит в столбцовом пространстве `A`),
  - либо решений **бесконечно много** (если уравнение совместно, но ранг недостаточный).

В моём вычислении значение `det(A)` выводится в ячейке выше. По нему я делаю вывод о разрешимости (существование/единственность решения).


```python
if np.isclose(detA, 0.0):
    print("Матрица A близка к вырожденной (det≈0). Уравнение может не иметь единственного решения.")
else:
    X = np.linalg.solve(A, B)
    print("Решение X найдено:")
    print(X)
```

```python
# Проверка: A @ X должно быть равно B (с учетом численной погрешности)
if not np.isclose(detA, 0.0):
    B_check = A @ X
    residual = B_check - B
    print("A @ X =
", B_check)
    print("B =
", B)
    print("Норма невязки ||A@X - B|| =", np.linalg.norm(residual))
```

### Вывод по проверке (как студент 2 курса)

После нахождения `X` я проверяю решение подстановкой: считаю `A @ X` и сравниваю с `B`.  
На практике **норма невязки** `||A@X - B||` получается очень маленькой, но не обязана быть строго 0, потому что:
- вычисления идут в **числах с плавающей точкой** (округления и погрешности),
- операции линейной алгебры используют численные методы, где ошибка накапливается.

Если невязка близка к 0 (например, порядка `1e-10…1e-12` или около того), я считаю, что решение найдено корректно, и уравнение выполнено с учётом численной точности.


## Задание 2. График функции и её производной

**Важно:** функция `f(x)` должна быть **вашей** (вариант преподавателя).

### 2.1 Полный график на области определения
Если есть точки разрыва (например, знаменатель = 0), область нужно разбить на интервалы.

Ниже реализован шаблон:
- вы задаёте список «запрещённых точек» (разрывы/асимптоты),
- строите график на каждом промежутке отдельно.

### Напишите своими словами (обязательно)
- Где у функции точки разрыва и почему?
- Какие особенности видно на графике (асимптоты, пересечения, экстремумы)?


```python
# ======== Настройки области построения (подберите под вашу функцию) ========
x_min, x_max = -10, 10

# Укажите точки разрыва/вертикальные асимптоты вашей функции (если есть).
# Для заглушки: x=1
break_points = [1.0]

# Параметры линии (сделайте отличными от умолчаний)
line_style = "--"     # например: "-", "--", "-.", ":"
line_width = 2.5
line_color = "tab:purple"

# Сетка значений на каждом интервале
N = 2000

def make_intervals(x_min, x_max, breaks, eps=1e-3):
    """Разбивает [x_min, x_max] на интервалы, исключая точки breaks (с небольшими окрестностями)."""
    points = [x_min] + sorted([b for b in breaks if x_min < b < x_max]) + [x_max]
    intervals = []
    for a, b in zip(points[:-1], points[1:]):
        if a == x_min and a in breaks:
            a = a + eps
        if b == x_max and b in breaks:
            b = b - eps
        # отступаем от разрыва по краям интервала
        if any(np.isclose(b, br) for br in breaks):
            b = b - eps
        if any(np.isclose(a, br) for br in breaks):
            a = a + eps
        if a < b:
            intervals.append((a, b))
    return intervals

intervals = make_intervals(x_min, x_max, break_points)
intervals
```

```python
plt.figure(figsize=(10, 5))

for a, b in intervals:
    xs = np.linspace(a, b, N)
    ys = f(xs)
    plt.plot(xs, ys, linestyle=line_style, linewidth=line_width, color=line_color)

# Вертикальные линии в точках разрыва/асимптот (если есть)
for bp in break_points:
    if x_min < bp < x_max:
        plt.axvline(bp, linestyle=":", linewidth=1.5)

plt.title("График функции f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.ylim(-20, 20)  # при необходимости подберите ограничение по y
plt.show()
```

### 2.2 Подписи, метки, настройки линий (как студент 2 курса)

- Диапазон по `x` (`x_min`, `x_max`) я выбираю так, чтобы на графике были видны **основные свойства** функции: пересечения с осями, экстремумы, асимптоты и характер роста/убывания.  
- Диапазон по `y` (`plt.ylim(...)`) иногда ограничиваю, потому что рядом с вертикальными асимптотами значения могут «улетать» в бесконечность и тогда «сплющивается» остальная часть графика.

Настройки линии в `plt.plot(xs, ys, ...)`:
- `color=...` задаёт **цвет** (я выбрал(а) не стандартный),
- `linestyle=...` задаёт **тип линии** (например `"--"` для пунктирной),
- `linewidth=...` задаёт **толщину линии**.

Вертикальные асимптоты/точки разрыва я показываю `plt.axvline(bp, linestyle=":", ...)` — получается вертикальная пунктирная линия в точке `x = bp`.

Ключевые точки (которые полезно отмечать на графике):
- пересечение с осью `Ox` (где `f(x)=0`),
- пересечение с осью `Oy` (значение `f(0)`, если определено),
- экстремумы (примерно где `f'(x)=0` и меняется знак производной),
- точки разрыва/асимптоты.


### 2.3 Производная и её график

Ниже:
- строим график `df(x)` (производной),
- при необходимости — делаем это на тех же интервалах, что и `f(x)`.

> Если у вас нет аналитической производной, можно использовать численную:
> 
`df_num(x) ≈ (f(x+h)-f(x-h))/(2h)`.


```python
# Если нужно — включите численную производную вместо df
USE_NUM_DERIV = False

def df_num(x, h=1e-5):
    return (f(x + h) - f(x - h)) / (2*h)

plt.figure(figsize=(10, 5))

for a, b in intervals:
    xs = np.linspace(a, b, N)
    ys = df_num(xs) if USE_NUM_DERIV else df(xs)
    plt.plot(xs, ys, linestyle="-.", linewidth=2.2)

for bp in break_points:
    if x_min < bp < x_max:
        plt.axvline(bp, linestyle=":", linewidth=1.5)

plt.title("График производной f'(x)")
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.grid(True)
plt.ylim(-20, 20)  # подберите под вашу функцию
plt.show()
```

### 2.4 Сопоставление графиков (как студент 2 курса)

Сравнивая графики `f(x)` и `f'(x)`, я использую базовые факты матанализа:

1) **Монотонность**  
- Если на интервале `f'(x) > 0`, то `f(x)` **возрастает**.  
- Если `f'(x) < 0`, то `f(x)` **убывает**.  
По графику производной видно, на каких промежутках она выше/ниже оси `Ox`, значит можно выделить интервалы возрастания и убывания.

2) **Экстремумы**  
Точки, где `f'(x)=0` (и при этом производная меняет знак), обычно соответствуют **локальным максимумам/минимумам** функции `f(x)`.  
- переход `+ → -` даёт максимум,  
- переход `- → +` даёт минимум.

3) **Асимптоты и разрывы**  
Если есть вертикальные асимптоты (точки разрыва), на графике `f(x)` видно, что функция стремится к `±∞` при приближении к этим `x`. Обычно вблизи таких точек и у производной поведение «резкое» (значения могут становиться очень большими по модулю), поэтому область определения приходится разбивать на интервалы.

В итоге я проверяю, что:
- интервалы монотонности по `f'(x)` совпадают с тем, что видно на графике `f(x)`,
- экстремумы на `f(x)` соответствуют нулям производной,
- учёт разрывов/асимптот согласуется с построением.


## Задание 3. Поиск локального минимума через `scipy.optimize.minimize`

Цель: найти точку локального минимума функции `f(x)` (из задания 2) запуском оптимизации из нескольких начальных точек.

> Для 1D-функции `minimize` работает, но часто удобнее `minimize_scalar`.
> Однако по условию используем **именно** `minimize()`.

### Что анализируем в результате
Типичный объект результата содержит:
- `x` — найденная точка минимума (оценка),
- `fun` — значение функции в этой точке,
- `success` — успешность,
- `message` — сообщение алгоритма,
- `nit` — число итераций,
- `nfev` — число вычислений функции,
- `jac` — оценка градиента (если доступно / если задавали).

Ниже шаблон запуска из 5 начальных точек.


```python
# Выберите 4–5 начальных точек (ориентируйтесь на графики из задания 2)
x0_list = [-8.0, -2.0, 0.5, 2.0, 8.0]  # <-- подберите под вашу функцию

# Ограничения (если надо): пример ограничений на x
# bounds = [(-10, 10)]  # для 1D minimize ожидает список кортежей
bounds = None

results = []
for x0 in x0_list:
    res = optimize.minimize(
        fun=lambda t: f(t[0]),          # minimize ожидает массив-параметр
        x0=np.array([x0], dtype=float),
        method="BFGS",                  # можно сравнить с 'Nelder-Mead'
        jac=(lambda t: np.array([df(t[0])])) if not USE_NUM_DERIV else None,
        bounds=bounds,
        options={"disp": False}
    )
    results.append((x0, res))

for x0, res in results:
    print("="*70)
    print("Стартовая точка x0 =", x0)
    print("success:", res.success)
    print("message:", res.message)
    print("x* =", res.x)
    print("f(x*) =", res.fun)
    print("nit =", getattr(res, "nit", None), " | nfev =", getattr(res, "nfev", None))
    print("jac =", getattr(res, "jac", None))
```

### Отчет по результатам (как студент 2 курса)

**1) Почему выбраны именно эти начальные точки?**  
Я выбрал(а) несколько стартовых точек `x0`, ориентируясь на график из задания 2:  
- часть точек взял(а) **вблизи** предполагаемого минимума (чтобы увидеть быструю сходимость),  
- часть — **далеко** (чтобы проверить, куда алгоритм придёт и насколько он устойчив),  
- если есть разрывы/асимптоты, то брал(а) точки **по разные стороны** от них, потому что оптимизация обычно работает только внутри одного «куска» области определения.

**2) Интерпретация результатов каждого запуска**  
Для каждого запуска `minimize()` я смотрю:
- `success`: `True` означает, что алгоритм считает поиск успешным (достигнут критерий остановки).  
- `message`: текстовое пояснение, почему остановились (например, достигнута точность, градиент мал, максимум итераций и т.п.).  
- `x*` (`res.x`): найденная точка (оценка аргумента минимума).  
- `fun` (`res.fun`): значение функции в найденной точке.  
- `nit`: сколько итераций сделал алгоритм.  
- `nfev`: сколько раз вычислялась функция.  
- `jac`: оценка градиента/производной (в 1D — это примерно `f'(x*)`). В точке минимума производная должна быть **близка к 0**.

**3) Влияние начальной точки**  
Да, начальная точка может сильно влиять: если у функции несколько локальных минимумов или есть разрывы, алгоритм обычно «падает» в ближайший (по траектории) минимум в пределах того интервала, где стартовали.

**4) Лучшее начальное приближение**  
Лучшим я считаю то `x0`, которое:
- привело к `success=True`,
- дало наименьшее `f(x*)` среди успешных запусков (если цели — именно минимум),
- потребовало меньше `nit`/`nfev` (быстрее сошлось),
- и результат согласуется с графиком (точка действительно выглядит как минимум).

**5) Для наиболее удачного запуска**  
По самому удачному `x0` я выписываю из вывода программы:
- `nit` — количество итераций,  
- `x*` — найденный аргумент минимума,  
- `fun` — значение минимума,  
- `jac` — производная/градиент в найденной точке (должен быть близок к 0),  
и сравниваю эту точку с графиком: должна находиться в «впадине» функции.

**6) Для неудачных запусков**  
Причины могут быть такие:
- старт слишком близко к разрыву/асимптоте (числа становятся огромными, алгоритм «путается»),  
- функция не гладкая или производная плохо считается,  
- алгоритм попал на плато/плохую область и остановился по лимиту итераций,  
- попал в другой локальный минимум (если он есть), что не совпадает с тем, что я ожидал(а).

**7) Общая оценка успешности**  
Если хотя бы один запуск дал `success=True`, разумное число итераций и точку, которая совпадает с графическим минимумом, то задачу считаю решённой успешно. Улучшить можно подбором `x0`, выбором метода (например, сравнить BFGS и Nelder-Mead) и аккуратным учётом разрывов (ограничения/разбиение области).
